```python
#write a fucntion to upload csv file in mongodb
def upload_csv_files(data_dir):
    #itearte the dir 
    for files in os.listdir(data_dir):
        if files.endswith('.csv'):
            collection_name=files.split('.')[0]#using csv name for collection
            collection=Database[collection_name]
            print(collection_name)

            Filepath=os.path.join(data_dir,files)
            print(Filepath)
            #read csv 
            df=pd.read_csv(Filepath)

            #remove extra spaces in columns name
            df.columns = (
             df.columns
              .astype(str)
             .str.strip()
             .str.replace('.', '_', regex=False)
             .str.replace('$', '', regex=False)
                 )
            df = df.where(pd.notnull(df), None)
            data=df.to_dict(orient='records')#we do this beacuse ,mongo db store data in dictionary that's why

            if data:
                collection.insert_many(data)
                print('len of data ',len(data))
            else:
                print('No data is Found')

```

```python
upload_csv_files(DATABASE_DIR)
```
